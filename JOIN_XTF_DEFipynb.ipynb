```python
import pandas as pd

df_principal = pd.read_excel('/content/30.01.25 Uic.xlsx', sheet_name='Export')
# subset_df = df_principal[df_principal['OpenchannelInteractionId'] == 1489]

# Cargar la hoja "Export" del archivo 30.01.25_Uic.xlsx
# df_principal = subset_df

# Cargar el archivo de tipificaciones (CSV)
df_tipificaciones = pd.read_excel('/content/20.02.25TipificacionesConversaciones.xlsx')
df_tipificaciones = df_tipificaciones.rename(columns={'ContactId': 'ContactId1'})

# 2️⃣ Cargar el archivo de Fichas CRM (suponiendo que es Excel)
df_fichas_crm = pd.read_excel('/content/20.02.25FichasCrm (2).xlsx')
df_fichas_crm = df_fichas_crm.rename(columns={'ContactId': 'ContactId2'})

# Realizar el LEFT JOIN
df_merged = pd.merge(
    df_principal,
    df_tipificaciones,
    how='left',
    left_on=['OpenchannelInteractionId', 'ContactId'],
    right_on=['id', 'ContactId1']
)

# Verificar el resultado
print(df_merged.head())

# Guardar el resultado en un nuevo archivo CSV (opcional)
df_merged.to_csv('resultado_left_join.csv', index=False)

# Convert 'createdAt' to datetime if it's not already
df_principal['createdAt'] = pd.to_datetime(df_principal['createdAt']) # Remove format to allow pandas to infer

# Format 'createdAt' to include only day, month, and year
df_principal['createdAt'] = df_principal['createdAt'].dt.strftime('%d-%m-%Y')

# Format 'Data/hora inici' to include only day, month, and year
# CORRECTED LINE: Remove format to allow pandas to infer the correct format
df_fichas_crm['Data/hora inici'] = pd.to_datetime(df_fichas_crm['Data/hora inici'])
df_fichas_crm['Data/hora inici'] = df_fichas_crm['Data/hora inici'].dt.strftime('%d-%m-%Y')

# Ensure df_merged1['createdAt'] also has the correct format before the final merge
# Apply the same formatting to df_merged1['createdAt'] and specify format
# using errors='coerce' to handle potential errors during conversion to datetime
# CHANGED LINE: Rename 'ContactId' to 'ContactId1' to match with df_fichas_crm
df_merged['createdAt'] = pd.to_datetime(df_merged['createdAt'], format='%d-%m-%Y', errors='coerce').dt.strftime('%d-%m-%Y')
df_merged1 = df_merged

# Realizar el LEFT JOIN sin renombrar ni crear columnas
df_merged_final = pd.merge(
    df_merged1,
    df_fichas_crm,
    how='left',
    left_on= ['ContactId1','createdAt'], # Use ContactId1 to match df_tipificaciones's ContactId1
    right_on=['ContactId2','Data/hora inici']
)


# Verificar el resultado
print(df_merged_final.head())

# Guardar el resultado en un nuevo archivo Excel
df_merged_final.to_excel('resultado_ficha_def.xlsx', index=False)
```
